```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# The Data
x = np.arange(-np.pi, 3*np.pi, 0.1)
y1 = np.cos(x)
y2 = np.cos(3/2*x)

# Create the figure and axes objects
fig,ax = plt.subplots(1, figsize=(8,6))
fig.suptitle('Sketching the graph')


# Show the grid lines as dark grey lines
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
ax.plot(x,y1, 'g:')
ax.plot(x,y2, 'b-')

```

```python
# The Data
x = np.arange(-np.pi, 4*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.sin(3*x)

# Create figure and axes
fig, ax = plt.subplots(1, figsize=(8, 6))

# Show grid lines as dark grey lines
plt.grid(b=True, which='major', color='#666666', linestyle='-')

ax.plot(x, y1, 'g:')
ax.plot(x, y2, 'b-')
```

```python
def radian(degree):
    return degree / 360 * 2 * np.pi

def degree(radian):
    return radian/np.pi*180

a = 35
b = 42
A = 25


constant = np.sin(radian(A))/a

B1 = degree(np.arcsin(b * constant))
B2 = 180 - B1

C1 = 180 - A - B1
C2 = 180 - A - B2

c1 = np.sin(radian(C1))/constant
c2 = np.sin(radian(C2))/constant

answer = 'B = %.1f or %.1f degree, C = %.1f or %.1f degree, c = %.1f or %.1f' % (B1, B2, C1, C2, c1, c2)
print(answer)
```
